```python
df = pd.read_csv(path+"Dataset/Italiano/dataset_BARD_recensioni_generate.csv", sep=';', encoding = "ISO-8859-1")
df.head()
```

```python
# source & destination languages
src = "it"
dst = "en"

task_name = f"translation_{src}_to_{dst}"
model_name = f"Helsinki-NLP/opus-mt-{src}-{dst}"

translator  = pipeline(task_name, model=model_name, tokenizer=model_name,max_length=600)
```

```python
def translate(df,llm):

  resuls = []
  ammount = df.shape[0]

  start = 0

  for i,r in df.iterrows():
    try:
      if i>= start:
        title = r["Titolo"]
        content = r["content"]

        title_eng = translator(title)[0]["translation_text"]
        content_eng = translator(content)[0]["translation_text"]

        l = list([r["ID"],title_eng,content_eng,  r["label"]])
        resuls.append(l)
        #----- Store the csv containing the results each iteration

        sys.stdout.write(
            "\r "+llm+" Done %i/%i                                " % (i, ammount))
        sys.stdout.flush()

        df_resuls = pd.DataFrame(resuls)
        df_resuls.columns = ["ID", "Title", "content", "label"]

        df_resuls.to_csv(path + "Dataset/dataset_" + llm + "_recensioni_generate_eng.csv", sep=";", index=None)
    except Exception as e:
      print(f"Index for restart: {i}")
      print(e)

    
```

```python
LLMs = ["BARD", "CLAUDE", "GPT", "LAMA"]

for llm in LLMs:
  df = pd.read_csv(path+"Dataset/Italiano/dataset_" + llm + "_recensioni_generate.csv", sep=';', encoding = "ISO-8859-1")
  translate(df,llm)
```
